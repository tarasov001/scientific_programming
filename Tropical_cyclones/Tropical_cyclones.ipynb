```python
def common_tc_dates(cyclone_dates_1, cyclone_dates_2):

    def make_set(dates):

        _31 = [1, 3, 5, 7, 8, 10, 12]
        _30 = [4, 6, 9, 11]
        _28 = [2]

        drop_chars = lambda s: ''.join([x for x in s if x.isdigit()])
        set_dates = set()

        for date in dates:
            tmp = date.split('/')
            if len(tmp) != 3:
                continue

            tmp = [drop_chars(x) for x in tmp]
            
            if len(tmp[0]) not in (1,2):
                continue
            elif len(tmp[1]) not in (1,2):
                continue
            elif len(tmp[2]) != 4:
                continue

            tmp = list(map(int, tmp))

            if tmp[1] > 0 and tmp[1] <= 12:
                    if tmp[2] >= 1982 and tmp[2] <= 2022:
                        if tmp[1] in _31 and tmp[0] > 0 and tmp[0] <= 31:
                            set_dates.add(f'{tmp[0]}.{tmp[1]}')
                        elif tmp[1] in _30 and tmp[0] > 0 and tmp[0] <= 30:
                            set_dates.add(f'{tmp[0]}.{tmp[1]}')
                        elif tmp[1] in _28 and tmp[0] > 0 and tmp[0] <= 28:
                            set_dates.add(f'{tmp[0]}.{tmp[1]}')
                        else:
                            continue
            else:
                continue

        return set_dates

    set_d1 = make_set(cyclone_dates_1)
    set_d2 = make_set(cyclone_dates_2)

    res = {
        "any_year": len(set_d1 | set_d2), 
        "both_years": len(set_d1 & set_d2), 
        "only_one_year": len(set_d1 ^ set_d2),
        "only_first_year": len(set_d1 - set_d2), 
        "only_second_year": len(set_d2 - set_d1), 
        "none_of_years": 365 - len(set_d1 | set_d2)
    }

    return res
```
